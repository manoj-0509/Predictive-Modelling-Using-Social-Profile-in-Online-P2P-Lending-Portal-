```python
#perform some feature engineering on the dataset. We'll create a new variable called loan_status_num that maps the 
#loan status to a numerical value.

# create a new variable called 'loan_status_num'
prosper_data['loan_status_num'] = prosper_data['LoanStatus'].map({'Completed': 1, 'Chargedoff': 0, 'Defaulted': 0})

# display the first 5 rows of the dataset with the new variable
prosper_data.head()
```

```python
#univariate analysis on the dataset.
# plot the distribution of loan amount
plt.figure(figsize=(10,6))
sns.distplot(prosper_data['LoanOriginalAmount'], kde=False)
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Loan Amount')
plt.show()
```

```python
#bivariate analysis 

# plot the relationship between loan amount and loan status
plt.figure(figsize=(10,6))
sns.boxplot(x='loan_status_num', y='LoanOriginalAmount', data=prosper_data)
plt.xlabel('Loan Status')
plt.ylabel('Loan Amount')
plt.title('Relationship between Loan Amount and Loan Status')
plt.show()
```

```python
#multivariate analysis

# plot the correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(prosper_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
