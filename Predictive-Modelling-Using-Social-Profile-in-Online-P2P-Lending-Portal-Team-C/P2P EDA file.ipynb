# Missing Values

```python
loan_df.isnull().mean()
```

```python
# columns with missing values more than 50 percent is deleted except Credit grade and closed date and also the columns not relevant to the data is dropped.
```

```python
loan_df.drop(["EstimatedEffectiveYield","EstimatedLoss","EstimatedReturn","ProsperRating (numeric)","ProsperRating (Alpha)"],axis = 1,inplace = True)
```

```python

loan_df.drop(['InvestmentFromFriendsCount', 'Recommendations' , 'LP_NonPrincipalRecoverypayments', 'LP_NetPrincipalLoss', 'LP_GrossPrincipalLoss','LP_CollectionFees','PercentFunded','LP_ServiceFees','LP_InterestandFees','LP_CustomerPrincipalPayments','LP_CustomerPayments','ProsperPaymentsOneMonthPlusLate','ProsperPaymentsLessThanOneMonthLate','OnTimeProsperPayments','TotalProsperPaymentsBilled','TotalProsperLoans'],axis = 1,inplace = True)
```

```python
loan_df.shape
```

```python
loan_df.drop(['LoanOriginationQuarter','LoanOriginationDate','LoanOriginalAmount','LoanNumber','LoanMonthsSinceOrigination','LoanFirstDefaultedCycleNumber','ProsperPrincipalOutstanding','ProsperPrincipalBorrowed','ListingCategory (numeric)','ListingCreationDate','ListingNumber','ListingKey'],axis = 1,inplace = True)
```

```python
loan_df.shape# now we are left with 37 columns.
```

```python
#Dealing with missing values.
#In CreditGrade around 70% values are missing..so filling null values with universal value.
loan_df["CreditGrade"].fillna(value = "others",inplace = True)
```

```python
loan_df.head()
```

```python
loan_df.fillna(value = 0)
```

```python
#Working for target variable.
#removing "current " row ffrom Loanstatus column
loan_df.drop(loan_df[loan_df["LoanStatus"] == "Current"].index, inplace = True)
```

```python
loan_df["LoanStatus"].value_counts()# Current row is removed.
```

```python
# Working on Target Variable with Closed days and Deliquent days.
loan_df["LoanStatus"] = loan_df["ClosedDate"].apply(lambda x: 1 if not pd.isnull(x) else 0)
```

```python
 loan_df["LoanStatus"].value_counts()
```

```python
loan_df['LoanCurrentDaysDelinquent'].value_counts()
```

```python
loan_df["LoanStatus"] = loan_df["LoanCurrentDaysDelinquent"].apply(lambda x: 1 if x>180 else 0)
```

```python
loan_df["LoanStatus"].value_counts()
```

# Detecting Outliers

```python
#defining a funtion plot_boxplot to check outliers in all column of data set
def plot_box(dfr,ft):#  ft are features.
    loan_df.boxplot(column=[ft])
    plt.grid(False)
    plt.show()
```
