```python
#removing outliers
def remove_outlier(col):
  Q1 = col.quantile(0.25)
  Q3 = col.quantile(0.75)
  IQR = Q3 - Q1
  lower_range = Q1 - 1.5 * IQR
  upper_range = Q1 + 1.5 * IQR
  return col[(col >= lower_range) & (col <= upper_range)]
```

# Feature Engineering

1. Fill the missing values

```python
#8.0 = NOT AVAILABLE
df_data.EmploymentStatus.fillna('8.0',inplace = True)

#checking value counts
df_data.EmploymentStatus.value_counts()
```

```python
#filling numerical data
df_data.BorrowerAPR = df_data['BorrowerAPR'].fillna(df_data['BorrowerAPR'].mean())
df_data.DebtToIncomeRatio = df_data['DebtToIncomeRatio'].fillna(df_data['DebtToIncomeRatio'].mean())
df_data.CreditScoreRangeUpper = df_data['CreditScoreRangeUpper'].fillna(df_data['CreditScoreRangeUpper'].mean())
df_data.CreditScoreRangeLower = df_data['CreditScoreRangeLower'].fillna(df_data['CreditScoreRangeLower'].mean())
df_data.TotalCreditLinespast7years = df_data['TotalCreditLinespast7years'].fillna(df_data['TotalCreditLinespast7years'].mean())
df_data.InquiriesLast6Months = df_data['InquiriesLast6Months'].fillna(df_data['InquiriesLast6Months'].mean())
df_data.CurrentDelinquencies = df_data['CurrentDelinquencies'].fillna(df_data['CurrentDelinquencies'].mean())
df_data.DelinquenciesLast7Years = df_data['DelinquenciesLast7Years'].fillna(df_data['DelinquenciesLast7Years'].mean())
df_data.PublicRecordsLast10Years = df_data['PublicRecordsLast10Years'].fillna(df_data['PublicRecordsLast10Years'].mean())
df_data.CreditGrade = df_data['CreditGrade'].fillna(df_data['CreditGrade'].mean())


#using median to fill data
df_data['ProsperRating (numeric)'].fillna(df_data['ProsperRating (numeric)'].median(), inplace = True)
df_data['ProsperRating (Alpha)'].fillna(df_data['ProsperRating (Alpha)'].median(), inplace = True)
df_data['ProsperScore'].fillna(df_data['ProsperScore'].median(), inplace = True)
df_data['EmploymentStatusDuration'].fillna(df_data['EmploymentStatusDuration'].median(), inplace = True)
df_data['IncomeVerifiable'].fillna(df_data['IncomeVerifiable'].median(), inplace = True)
df_data['AvailableBankcardCredit'].fillna(df_data['AvailableBankcardCredit'].median(), inplace = True)
df_data['TotalInquiries'].fillna(df_data['TotalInquiries'].median(), inplace = True)


#filling categorical data
df_data['Occupation'].fillna(df_data['Occupation'].mode()[0], inplace = True)
```
